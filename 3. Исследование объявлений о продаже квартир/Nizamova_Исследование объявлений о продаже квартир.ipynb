## Предобработка данных

```python
df['balcony'] = df['balcony'].fillna(0)   # Заменим пропуски на 0 в аппартаментах и балконах.
df['is_apartment'] = df['is_apartment'].fillna(0)
```

```python
df['ceiling_height'].sort_values().unique() # Какие уникальные значения есть в столбце?
```

```python
df['ceiling_height'] = df['ceiling_height'].fillna(value=(df['ceiling_height'].median()))  # Заменила пропуски медианой
# df.loc[(df['ceiling_height'] < 2.5) & (df['ceiling_height'] > 4.3)] = 2.65 
# Заменила все "не стандартные" значения на медиану- формула работает через раз :(. Выбросы удалила по другой схеме.
```

На основании СНиП 31-01-2003 Здания жилые многоквартирные: 

5.8 Высота (от пола до потолка) жилых помещений и кухни (кухни-столовой) в климатических районах IА, IБ, IГ, IД и IVA должна быть не менее 2,7 м, а в других климатических районах - не менее 2,5 м.
У Санкт-Петербурга климатический район IIB. Значит берем минималку 2,5 м. А максималка обнаружена в жилье бизнес-класса= 4,3м. Остальное выбросы. Пропущенные значения можно заменить медианой.

```python
# Высокое значение стандартного отклонения, взяла медиану.
df['days_exposition'] = df['days_exposition'].fillna(value=(df['days_exposition'].median()))
# Преобразуем в формат даты
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
# Цену в тысячи рублей (укрупним)
df['last_price'] = df['last_price'] / 1000
# Пропуски заменяю на нули.
df['cityCenters_nearest'] = df['cityCenters_nearest'].fillna(value=0)

df.describe()
```

## Расчёты и добавление результатов в таблицу

### Посчитайте и добавьте в таблицу цену квадратного метра.

```python
df['price_m2'] = (df['last_price'] / df['total_area']).astype(int)  # Преобразуем в целое число.
```

### Посчитайте и добавьте в таблицу день недели, месяц и год публикации объявления

```python
df['weekdays'] = pd.DatetimeIndex(df['first_day_exposition']).weekday    #  Начну со дня недели
df['months'] = pd.DatetimeIndex(df['first_day_exposition']).month  # Столбец по месяцам
df['years'] = pd.DatetimeIndex(df['first_day_exposition']).year  # Столбец по годам
```

### Посчитайте и добавьте в таблицу этаж квартиры; варианты — первый, последний, другой

```python
def apartment_floor(row):     
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'

df['apartment_floor'] = df.apply(apartment_floor, axis=1)
df.head()
```

### Посчитайте и добавьте в таблицу соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
df_area = df[['total_area', 'living_area', 'kitchen_area', 'studio', 'rooms', 'open_plan']].reset_index(drop=True) 
df_area # Для наглядности работы с площадями создадим новый датафрей
```

Заполняем пропуски в площадях жилых помещений и кухни, вычитая от общей площади. Конечно в реальности, необходимы площади по мокрым зонам (санузлы, ванные, постирочные) и нежилым (коридорам, прихожей, кладовой, гардеробной и других подсобных помещений), тогда результат будет более реалистичным.

```python
df_area['living_area'] = df_area['living_area'].fillna(value=(df_area['total_area'] - df_area['kitchen_area'])) 
df_area['kitchen_area'] = df_area['kitchen_area'].fillna(value=(df_area['total_area'] - df_area['living_area'])) 
df_area['kitchen_area'] = df_area['kitchen_area'].fillna(0) # Заполняем оставшиеся пропуски на нуль, что характерно для студий.
                                                            # Площадь кухни включена в общую площадь
df_area['living_area'] = df_area['living_area'].fillna(0)   # Заполняем на нуль по жилым помещениям, для свободной планировки  
df_area.info()
```

```python
df_area['ratio_residential'] = df_area['living_area'] / df_area['total_area']    # соотношение жилой и общей площади, 
df_area['ratio_kitchen'] = df_area['kitchen_area'] / df_area['total_area']       # а также отношение площади кухни к общей
df_area.head()
```

Нашла пропущенные данные на кухне и жилой площади, путем вычитания и заполения нулем (т.к. кухни в студиях и квартирах свободной планировки включены в площадь).

### Уберите редкие и выбивающиеся значения. Опишите, какие особенности обнаружили.

```python
# С помощью функции уберем выбросы
def quartile_df(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    dfq = df.loc[(df[column] < q3 + 1.5*iqr) & (df[column] > q1 - 1.5*iqr), column]
    return dfq
# Выделим столбцы с выбросами, которые необходимо обработать
list_of_emissions = ['last_price', 'total_area', 'kitchen_area', 'living_area', 'rooms', 'ceiling_height']
for col in list_of_emissions:
    df[col] = quartile_df(df, col)
df.describe()
```
